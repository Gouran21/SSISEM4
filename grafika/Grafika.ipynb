```python
from PIL import Image, ImageDraw

# Ładowanie obrazka
input_image = Image.open("input.png")
input_pixels = input_image.load()

# MAcierz blur
box_kernel = [[1 / 9, 1 / 9, 1 / 9],
              [1 / 9, 1 / 9, 1 / 9],
              [1 / 9, 1 / 9, 1 / 9]]
gaussian_kernel = [[1 / 256, 4  / 256,  6 / 256,  4 / 256, 1 / 256],
                   [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
                   [6 / 256, 24 / 256, 36 / 256, 24 / 256, 6 / 256],
                   [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
                   [1 / 256, 4  / 256,  6 / 256,  4 / 256, 1 / 256]]

# wybór jądra
kernel = box_kernel

# środek macierzy
offset = len(kernel) // 2

# tworzenie obrazka wyściowego
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

# proces blura
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * kernel[a][b]
                acc[1] += pixel[1] * kernel[a][b]
                acc[2] += pixel[2] * kernel[a][b]

        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))
    
output_image.save("output.png")
```

```python
from PIL import Image, ImageDraw
from math import sqrt

# Ładowanie obrazka
input_image = Image.open("input.png")
input_pixels = input_image.load()

# Obliczanie średniej ostrości koloru pixeli
intensity = [[sum(input_pixels[x, y]) / 3 for y in range(input_image.height)] for x in range(input_image.width)]
# Jądro sobel
kernelx = [[-1, 0, 1],
           [-2, 0, 2],
           [-1, 0, 1]]
kernely = [[-1, -2, -1],
           [0, 0, 0],
           [1, 2, 1]]

# Tworzenie borazka wyjsciowego
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

# Proces
for x in range(1, input_image.width - 1):
    for y in range(1, input_image.height - 1):
        magx, magy = 0, 0
        for a in range(3):
            for b in range(3):
                xn = x + a - 1
                yn = y + b - 1
                magx += intensity[xn][yn] * kernelx[a][b]
                magy += intensity[xn][yn] * kernely[a][b]

        # Rysowanie w czerni i bieli 
        color = int(sqrt(magx**2 + magy**2))
        draw.point((x, y), (color, color, color))
    
output_image.save("output.png")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Ładowanie obrazka
img0 = cv2.imread("input.png")

# Zamiana na odcień szarości
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# Usuwanie szumów
img = cv2.GaussianBlur(gray,(3,3),0)

# Macierz Laplaca
laplacian = cv2.Laplacian(img,cv2.CV_64F)

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#harris
import cv2
import numpy as np

filename = "input.png"
img = cv2.imread(filename)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)

dst = cv2.dilate(dst,None)

img[dst>0.01*dst.max()]=[0,0,255]

cv2.imshow('dst',img)
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```
